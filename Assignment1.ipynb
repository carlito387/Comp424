```python
#Create k-ary tree object
class Node : 
    #Create a new tree node 
    def __init__(self ,sentence, cost, depth): 
        self.sentence = sentence
        self.cost = cost
        self.depth = depth
        self.child = []
        
#Creates the desired tree
def create_tree(limit_depth):
    if limit_depth>6:
        print("TREE ASKED IS TOO BIG")
        return
    words=[("",0)]*5
    words[0]=("cat",2); words[1]=("mat",2); words[2]=("on",1); words[3]=("sat",2); words[4]=("the",2)
    root = Node("_",0,0) 
    for i in range(5):
        (word_1,consonants_1)=words[i];
        root.child.append(Node(word_1,consonants_1,1));
        if limit_depth!=1:
            for j in range(5):
                (word_2,consonants_2)=words[j];
                root.child[i].child.append(Node(word_1+" "+word_2,consonants_1+consonants_2,2));
                if limit_depth!=2:
                    for k in range(5):
                        (word_3,consonants_3)=words[k];
                        root.child[i].child[j].child.append(Node(word_1+" "+word_2+" "+word_3,consonants_1+consonants_2+consonants_3,3));
                        if limit_depth!=3:
                            for l in range(5):
                                (word_4,consonants_4)=words[l];
                                root.child[i].child[j].child[k].child.append(Node(word_1+" "+word_2+" "+word_3+" "+word_4,consonants_1+consonants_2+consonants_3+consonants_4,4));
                                if limit_depth!=4:
                                    for n in range(5):
                                        (word_5,consonants_5)=words[n];
                                        root.child[i].child[j].child[k].child[l].child.append(Node(word_1+" "+word_2+" "+word_3+" "+word_4+" "+word_5,consonants_1+consonants_2+consonants_3+consonants_4+consonants_5,5));
                                        if limit_depth!=5:
                                            for m in range(5):
                                                (word_6,consonants_6)=words[m];
                                                root.child[i].child[j].child[k].child[l].child[n].child.append(Node(word_1+" "+word_2+" "+word_3+" "+word_4+" "+word_5+" "+word_6,consonants_1+consonants_2+consonants_3+consonants_4+consonants_5+consonants_6,6));
    return root

#Prints the tree
def printer(root):
    if root is None: 
        return
    for i in range(root.depth):
        print("~~~~~", end="")
    print(root.sentence)
    for child in root.child:
        printer(child)
        
#Prints the node in order when the tree is traversed by breadth first search, using the traditionnal queue method
def print_bfs(root):
    if root is None:
        return
    queue=[]
    queue.append(root)
    while queue:
        s=queue.pop(0)
        for child in s.child:
            queue.append(child)
        print(s.sentence)
        if s.sentence=="the cat sat":
            break

#Prints the node in order when the tree is traversed by uniform cost search
def print_ucs(root, limit_cost):
    if root is None:
        return
    ordered_nodes=[(root.sentence,root.cost)]
    queue=[]
    queue.append(root)
    i=0
    #BFS for putting all the elements of the tree in the list ordered_nodes
    while queue:
        s=queue.pop(0)
        for child in s.child:
            queue.append(child)
        #Handle the root case
        if i!=0:
            ordered_nodes.append((s.sentence,s.cost))
        i=1
    #Sort alphabetically ordered_nodes
    ordered_nodes=sorted(ordered_nodes, key=itemgetter(0));
    #Iterate through it many times, and output in order the words with the same cost, alphabetically
    for goal_cost in range(limit_cost+1):
        for (sentence,cost) in ordered_nodes:
            if cost==goal_cost:
                print(sentence)
                #Stop if you meet a solution
                if sentence=="the cat sat":
                    return
                    
#Prints the node in order when the tree is traversed by iterative deepening search
def print_ids(root,limit_depth):
    if root is None:
        return
    ordered_nodes=[(root.sentence,root.depth)]
    queue=[]
    queue.append(root)
    i=0
    #BFS for putting all the elements of the tree in the list ordered_nodes
    while queue:
        s=queue.pop(0)
        for child in s.child:
            queue.append(child)
        #Handle the root case
        if i!=0:
            ordered_nodes.append((s.sentence,s.depth))
        i=1
    #Sort alphabetically ordered_nodes
    ordered_nodes=sorted(ordered_nodes, key=itemgetter(0));
    #Iterate through it many times, and output in order the words with the same depth, alphabetically
    for goal_depth in range(limit_depth+1):
        for (sentence,depth) in ordered_nodes:
            if depth==goal_depth:
                print(sentence)
                #Stop if you meet a solution
                if sentence=="the cat sat":
                    return

root=create_tree(6)
#printer(root)
#print_bfs(root)
#print_ucs(root,12)
#print_ids(root,6)
```

Note that we used the following formula to compute the standard deviations:
$$S=(\frac{n\times\sum_{i=1}^nx_i^2-(\sum_{i=1}^nx_i)^2}{n\times(n-1)})^{\frac{1}{2}}$$

## QUESTION 3) a)

```python
def f1(x,y):
    return math.sin(x/2)+math.cos(2*y)
def f2(x,y):
    return -abs(x-2)-abs(y/2+1)+3
#Outputs the set of neighbours of a point (x,y)
def neighbours(x,y,step):
    return [(x-step,y-step), (x,y-step), (x+step,y-step), (x-step,y), (x+step,y), (x-step, y+step), (x,y+step), (x+step, y+step)]
#Outputs the neighbour with the highest value under f
def better_neighbour(x,y,best,step,function):
    for (x_new,y_new) in neighbours(x,y,step):
        if (0<=x_new) and (x_new<=10) and (0<=y_new) and (y_new<=10) and function(x_new,y_new)>best:
            best=function(x_new,y_new); (x,y)=(x_new,y_new)
    return(x,y,best)

for function in [f1,f2]:
    if (function==f1): print("FUNCTION 1:") 
    else: print("FUNCTION 2:")
    for step in [0.01,0.05,0.1,0.2]:
        sum=0; average=0; squared_average=0; squared_sum=0
        for i in range(0,100):
            iterator=0
            #Initialize a random point, and take its neighbour with the highest value under f
            (x,y)=(10*rand.uniform(0,1),10*rand.uniform(0,1)); best=function(x,y)
            (x_new,y_new,best_new)=better_neighbour(x,y,best,step,function)
            #Repeat while there is no neighbour with a higher value
            while best!=best_new:
                best=best_new; x=x_new; y=y_new
                (x_new,y_new,best_new,)=better_neighbour(x,y,best,step,function)
                iterator=iterator+1
            average=average+best; sum=sum+iterator; squared_average=squared_average+pow(best,2); squared_sum=squared_sum+pow(iterator,2) 
        average_conv=sum/100; average_val=average/100; standard_val=math.sqrt((100*squared_average-pow(average,2))/(100*99)); standard_conv=math.sqrt((100*squared_sum-pow(sum,2))/(100*99))
        print("step:",step,"average convergence:",average_conv,"standard deviation convergence",standard_conv,"average value:",average_val,"standard deviation value:", standard_val)
```
